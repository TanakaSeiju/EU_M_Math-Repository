```python
import numpy as np
import matplotlib.pyplot as plt
import math

N_in = 0 
N_out = 0

#試行回数
N = 10000

#乱数の指定
x = np.random.uniform(0.0, 1.0, N)
y = np.random.uniform(0.0, 1.0, N)

for i in np.hypot(x, y):
  if  i < 1:
       N_in += 1
  else:
       N_out += 1

#πの近似
PI = N_in / N *4

print("円に入った数:",N_in)

c1 = plt.Circle((0, 0), radius=1, fc="None", ec="r", linewidth=2, color = "black")
ax = plt.gca()
ax.add_patch(c1)
plt.axis("scaled")
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("PLOT")

plt.scatter(x, y, marker=".", color = "green", label = "POINT")
plt.show()
```

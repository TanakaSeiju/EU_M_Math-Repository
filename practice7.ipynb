```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00312/dow_jones_index.zip'

r = requests.get(url, stream = True)

z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()
dow_jones_index = pd.read_csv("dow_jones_index.data", sep = ",")
dow_jones_index.head()
```

```python
dow_jones_index.info
```

```python
dow_jones_index.date = pd.to_datetime(dow_jones_index.date)

dol = lambda x: str(x).replace("$", " ")

dow_jones_index.open = pd.to_numeric(dow_jones_index.open.map(dol))
dow_jones_index.high = pd.to_numeric(dow_jones_index.high.map(dol))
dow_jones_index.low = pd.to_numeric(dow_jones_index.low.map(dol))
dow_jones_index.close = pd.to_numeric(dow_jones_index.close.map(dol))
dow_jones_index.next_weeks_open = pd.to_numeric(dow_jones_index.next_weeks_open.map(dol))
dow_jones_index.next_weeks_close = pd.to_numeric(dow_jones_index.next_weeks_close.map(dol))

dow_jones_index.head()
```

```python
dow_jones_re_index = dow_jones_index.set_index(["date", "stock"])
dow_jones_re_index_unstoak = dow_jones_re_index.unstack()
date1 = dow_jones_re_index_unstoak["close"]
date1.describe()
```

```python
corr_mat = date1.corr()
corr_mat
```

```python
sns.heatmap(corr_mat)
```

```python
max_corr = 0
stock1 = " "
stock2 = " "

for i in range(0, len(corr_mat)):
  print(
      corr_mat[i:i+1].unstack().sort_values(ascending = False)[[1]].idxmax()[1],
      corr_mat[i:i+1].unstack().sort_values(ascending = False)[[1]].idxmax()[0],
      corr_mat[i:i+1].unstack().sort_values(ascending = False)[[1]][0]
  )
  if max_corr < corr_mat[i:i+1].unstack().sort_values(ascending = False)[[1]][0]:
    max_corr = corr_mat[i:i+1].unstack().sort_values(ascending = False)[[1]][0]
    stock1 = corr_mat[i:i+1].unstack().sort_values(ascending = False)[[1]].idxmax()[1]
    stock2 = corr_mat[i:i+1].unstack().sort_values(ascending = False)[[1]].idxmax()[0]
  
print("[Max_crr]:", max_corr)
print("[stock1]:", stock1)
print("[stock2]:", stock2)
```

```python
date2 = date1[[stock1, stock2]]
date2.plot(subplots = True, grid = True)
plt.grid(True)
```

```python
date3 = np.log(date1/date1.shift(1))
 
date3_max = date3.std().idxmax()
date3_min = date3.std().idxmin()

print("max:", date3_max)
print("min:", date3_min)

date3[date3_max].plot()
date3[date3_min].plot()
plt.ylabel("log ratio")
plt.legend()
```

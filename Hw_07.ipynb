```python
#UAV
#

import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import codecs

with codecs.open("LOG119.csv", "r", "Shift-JIS", "ignore") as file:
    df = pd.read_table(file, delimiter=",")

date1 = df[[ "latitude", "longitude"]]
date2 = date1.dropna()
date3 = date2.drop(0)

fig = go.Figure(go.Scattermapbox(
    mode = "markers+lines",
    lon = date3["longitude"],
    lat = date3["latitude"],
    marker = {"size": 10, "color" : "Blue"}))


fig.update_layout(
    margin ={"l":0,"t":0,"b":0,"r":0},
    mapbox = {
        "style": "stamen-terrain",
        "center": {"lon": 136.854642, "lat": 34.954571 },
        "zoom": 12},
    showlegend = False)


fig.show()
```

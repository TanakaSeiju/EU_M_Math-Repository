```python
import numpy as np
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y, color = "red")
plt.plot(x, z, color = "blue")

plt.xlabel("X")
plt.ylabel("red_sinx blue_cosx")
plt.grid(True)
```

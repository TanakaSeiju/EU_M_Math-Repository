```python
import numpy as np
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt

def hist(N):
  plt.subplot(2, 1, 1)
  a = np.random.uniform(0.0, 1.0, N)
  plt.hist(a)
  plt.title("No1")
  plt.grid(True)

  plt.subplot(2, 1, 2)
  b = np.random.uniform(0.0, 1.0, N)
  plt.hist(b)
  plt.title("No2")
  plt.grid(True)

  plt.tight_layout()

hist(1000);
```

```python
hist(100);
```
